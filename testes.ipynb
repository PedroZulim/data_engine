```python
df2['firt name'] = df['firt name'].str.replace('#','e').str.replace('@','a').str.replace('!','i').str.replace('$','o').str.replace('%','a').str.replace('-',' ')
```

```python
df2['last name'] = df['last name'].str.replace('ðŸ˜€','o').str.replace('ðŸ˜‚','u').str.replace('ï¿½','i').str.replace('$','e').str.replace('3','a').str.capitalize()
```

```python
def replace_special_chars_after_at(text):
    if isinstance(text, str): 
        parts = text.split('@')
        if len(parts) == 2:
            cleaned_part = re.sub(r'[^a-zA-Z0-9]', '.', parts[1])
            return parts[0] + '@' + cleaned_part
    return text
```

```python
df2['email'] = df['email'].replace(np.nan, 'UNDEFINED').str.replace('$','@').str.replace('#','@').str.replace('!','@').str.replace('tmall.','hotmail.').str.strip()
```

```python
df2['email'] = df['email'].apply(replace_special_chars_after_at)
```

```python
def clean_email(text):
  if text != "UNDEFINED":
    local_part, domain_part = text.split('@')
    cleaned_local_part = re.sub(r'(.)\1+', r'\1', local_part)
    return f'{cleaned_local_part}@{domain_part}'
  return text
```

```python
df2['email'] = df['email'].apply(clean_email)
```
