```python
def replace_special_chars_after_at(text):
    if isinstance(text, str): 
        parts = text.split('@')
        if len(parts) == 2:
            cleaned_part = re.sub(r'[^a-zA-Z0-9]', '.', parts[1])
            return parts[0] + '@' + cleaned_part
    return text
```

```python
df2['email'] = df2['email'].str.replace('$','@').str.replace('#','@').str.replace('!','@').str.replace('tmall.','hotmail.').str.strip()
```

```python
df2['email'] = df2['email'].apply(replace_special_chars_after_at)
```

```python
def clean_email(text):
  if text == "UNDEFINED" or pd.isnull(text):
    return text
  local_part, domain_part = text.split('@')
  cleaned_local_part = re.sub(r'(.)\1+', r'\1', local_part)
  return f'{cleaned_local_part}@{domain_part}'
  
```

```python
df2['email'] = df2['email'].apply(clean_email)
```
